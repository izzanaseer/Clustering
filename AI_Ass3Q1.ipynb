```python
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
from sklearn_extra.cluster import KMedoids
import matplotlib.pyplot as plt
from sklearn.metrics import silhouette_score

# Generate random data points
X, _ = make_blobs(n_samples=300, centers=4, cluster_std=0.60, random_state=0)

# K-means clustering
kmeans = KMeans(n_clusters=4)
kmeans.fit(X)
kmeans_labels = kmeans.labels_
kmeans_centers = kmeans.cluster_centers_

# K-medoids clustering
kmedoids = KMedoids(n_clusters=4)
kmedoids.fit(X)
kmedoids_labels = kmedoids.labels_
kmedoids_medoids = kmedoids.cluster_centers_

# Plotting results
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.scatter(X[:, 0], X[:, 1], c=kmeans_labels, cmap='viridis')
plt.scatter(kmeans_centers[:, 0], kmeans_centers[:, 1], marker='o', c='red', s=200, label='Centroids')
plt.title('K-means Clustering')
plt.legend()

plt.subplot(1, 2, 2)
plt.scatter(X[:, 0], X[:, 1], c=kmedoids_labels, cmap='viridis')
plt.scatter(kmedoids_medoids[:, 0], kmedoids_medoids[:, 1], marker='o', c='red', s=200, label='Medoids')
plt.title('K-medoids Clustering')
plt.legend()

plt.show()

# Calculate silhouette scores for k-means and k-medoids
silhouette_score_kmeans = silhouette_score(X, kmeans_labels)
silhouette_score_kmedoids = silhouette_score(X, kmedoids_labels)

print(f"Silhouette Score (K-means): {silhouette_score_kmeans}")
print(f"Silhouette Score (K-medoids): {silhouette_score_kmedoids}")
```
